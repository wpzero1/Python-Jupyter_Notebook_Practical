```python
#Matplotlib 실습

import matplotlib.pyplot as plt

#bar chart

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

#막대 너비 기본값은 0.8이므로, 막대가 가운데로 올 수 있도록 왼쪽 좌표에 0.1 더하기
xs = [i + 0.1 for i, _ in enumerate(movies)]

#왼편으로부터 x축의 위치가 xs이고 높이가 num_oscars인 막대 그리기
plt.bar(xs, num_oscars)
plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")

#막대의 가운데에 오도록 영화 제목 레이블 달기
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```

```python
from collections import Counter 
plt.style.use('ggplot')
#막대 그래프를 활용한 히스토그램

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade:grade // 10*10
histogram = Counter(decile(grade) for grade in grades) #counter는 연속된 값을 int와 유사한 객체로 변환. 빈도 연결용도

plt.bar([x - 4 for x in histogram.keys()], #각 막대를 왼쪽으로 4만큼 옮기고
         histogram.values(),                #높이설정
         8)                                 #너비는 8 (plt.bar의 3번째 인자는 너비.)
plt.axis([-5, 105, 0, 5]) #x축은 -5부터 105, y축은 0부터 5

plt.xticks([10 * i for i in range(11)]) #x축의 레이블은 0부터 10단위로 100
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

```python
# 선 그래프

variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error = [x+y for x,y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

#한 차트에 여러개의 series를 그리기 위해
#plt.plot을 여러번 호출 가능

plt.plot(xs, variance, 'g-', label='variance') #초록색 실선
plt.plot(xs, bias_squared, 'r-.', label='bias^2') #붉은색
plt.plot(xs, total_error, 'b:', label='total error') #파란색

#각 series에 label을 미리 달아놨기 때문에 범례를 어렵지 않게 그릴 수 있다. loc=9는 top center를 의미
plt.legend(loc=9) #숫자를 바꿔보면 위치가 달라진다
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

```python
#scatterplots

# 두 변수간의 연관관계.
# 각 사용자의 친구 수와 그들의 매일 사이트에서 체류하는 시간 사이의 연관성

friends = [70,65,72,63,71,64,60,64,67]
minutes = [175,170,205,120,220,130,105,145,190]
labels=['a','b','c','d','e','f','g','h','i']

plt.scatter(friends, minutes)

#각 포인트에 레이블 달기
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy=(friend_count, minute_count), #label을 데이터 포인트 근처에 두되 약간 떨어지게
                 xytext=(5,-5),
                 textcoords='offset points')
```
