```python
#Matplotlib : 주로 2차원의 데이터를 시각화하기 위한 서드파티 패키지. 자세한 그리기 설정 가능하다는 장점

#모듈 불러오기
import matplotlib.pyplot as plt
```

```python
#피겨와 서브플롯

#피겨는 서브플롯을 작성하는 영역, 서브플롯은 그래프를 작성하기 위한 영역

#add_subplot() 메서드로 서브플롯 배치

#피겨 생성
fig = plt.figure()

#피겨 안에 서브플롯 3개 배치
ax1 = fig.add_subplot(221) #2행2열 1번
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)

plt.show()
```

```python
#서브 플롯 번호 할당 순서 확인

#피겨 생성
fig = plt.figure()

#피겨 안에 서브플롯 3개 배치
ax1 = fig.add_subplot(221) #2행2열 1번
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)

#번호 기술
for i, ax in enumerate([ax1, ax2, ax3], start=1):
    txt = 'ax{0}\n(22{0})'.format(i)
    ax.text(0.2, 0.4, txt, fontsize=24)
    
plt.show()
```

```python
#subplot() 함수로 서브플롯배치하기

# 피겨 오브젝트 작성과 서브플롯 배치를 동시에 실행
flg, axes = plt.subplots(2, 2)
print(type(axes), axes)
plt.show()
```

```python
#행렬로 서브플롯 위치 지정

#1행 2열째 서브플롯에 서브플롯 타이틀 지정
fig, axes = plt.subplots(2,2)
axes[0,1].set_title('Subplot 0-1')
plt.show()
```

```python
#꺾은선 그래프
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1, 3]) #이부분이 x축, y축 설정. 리스트형 오브젝트가 넘겨진다.

plt.show()
```

```python
#x,y값을 지정한 꺾은선 그래프

fig = plt.figure()
ax = fig.add_subplot(111)

x = [0,2,4]
y = [0,4,2]

ax.plot(x,y)
plt.show()
```

```python
#선이 여러개

fig = plt.figure()
ax = fig.add_subplot(111)

x = [0,2,4]
y1 = [0,4,2.5]
y2 = [4,0,1.5]

#2개의 선
ax.plot(x,y1)
ax.plot(x,y2)

plt.show()
```

```python
# 꺾은선 그래프 활용

import os
import pandas as pd

anime_stock_returns_csv = './anime/anime_stock_returns.csv'

df = pd.read_csv(anime_stock_returns_csv, index_col=0, parse_dates=['Date'])
df.head()
```

```python
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(111)

#데이터와 범례 지정
ax.plot(df.index, df['TOEI ANIMATION'], label='TOEI ANIMATION')
ax.plot(df.index, df['IG Port'], label='IG Port')

#타이틀, 축 레이블 설정
ax.set_title('주가등락률 2년간 추이')
ax.set_ylabel('주가등락률')
ax.set_xlabel('년월')
```

```python
#두개의 축을 가진 그래프 : Axes.twinx()

t4816_csv = './anime/4816.csv'
df = pd.read_csv(t4816_csv, index_col=0, parse_dates=['Date'])

fig = plt.figure(figsize = (10,4))
ax1 = fig.add_subplot(111)

ax1.plot(df.index, df['Close'], color='b', label='주가')

# X축을 공유해서 Y축을 2개 사용하는 설정
ax2 = ax1.twinx()
ax2.bar(df.index, df['Volume'], color = 'g', label = '거래총액', width=2)

# 축과 축레이블 설정
ax1.set_yticks([i * 2000 for i in range(5)])
ax1.set_ylabel('주가')
ax2.set_yticks([i * 50000 for i in range(5)])
ax2.set_ylabel('거래총액')
ax1.set_xlabel('년월')

# 그래프 타이틀 설정
ax1.set_title('주가와 거래초액')

#범례설정
ax1.legend(loc=1)
ax2.legend(loc=2)

plt.show()
```
