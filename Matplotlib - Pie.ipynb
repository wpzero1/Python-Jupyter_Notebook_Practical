```python
#원그래프 : Axes.pie()

import matplotlib.pyplot as plt
plt.style.use('ggplot')

labels = ['자전거', '버스', '차']
sizes = [25, 40, 35]

fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

#원 그래프 그리기
ax.pie(sizes, labels=labels)

plt.show()
```

```python
#원그래프와 축의 관계 그리기


fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

ax.pie(sizes, labels=labels, radius = 0.9, frame=True)
ax.text(-0.3, 0, '(0,0)', fontsize=9)

plt.show()
```

```python
#원그래프 서식 설정

fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

#부채꼴 서식 설정용 사전
wprops = {'edgecolor':'black', 'linewidth':2}

#텍스트 서식 설정용 사전
tprops = {'fontsize':18}

ax.pie (sizes,
        explode=(0, 0.05, 0), #요소의 분리 여부. 해당하는 차례에 0.05
        labels=labels, 
        autopct='%1.0f%%', #수치 레이블 서식
        pctdistance=0.5, #수치 레이블 위치
        shadow=False, #배경의 표시, 비표시. 논리값 지정
        labeldistance=1.35,
        startangle=90,
        radius=0.3,
        counterclock=False, #표시순서 : False - 시계, True(기본값) : 반시계
        wedgeprops=wprops, # 각 요소 서식
        textprops=tprops, #텍스트 서식
        center=(0.5, 0.5), #중심좌표, 튜플형
        frame=True) #축, 테두리선 유무 논리값

plt.show()
```
